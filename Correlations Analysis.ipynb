# To explore how the different variables are related to each other.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
power_lifting = pd.read_csv("C:/Users/evjoh/Downloads/power_lifting.csv")
```

```python
power_lifting.head()
```

## Drop Non-Continuous Variables for Analysis

```python
power_lifting1 = power_lifting.drop(['Name', 'Sex', 'Equipment', 'Division'], axis=1)
power_lifting1.head()
```

```python
power_lifting1.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
# Correlation matrix using seaborn package:

sns.heatmap(power_lifting1.corr(), annot=True)
```
